```python
def calcute_word_log_prob_per_class(frequences1,n_tokens1,n_unique_tokens1):
    for i in list(frequences1.keys())[:len(frequences1)]:
        #smoothing 
        frequences1[i] = np.log10((frequences1[i]+1)/(n_tokens1+n_unique_tokens1))
    return frequences1
```

```python
def calcute_log_prior(tokens_list_per_class,n_tokens):
    prior_log_list=[]
    for i in range(0,len(tokens_list_per_class)):
        prior_log_list.append( np.log10(tokens_list_per_class[i]/n_tokens))
    return prior_log_list
```

```python
def predict(split_list):
    list_maxIndex=[]
    for l in split_list:
        predict_list=[]
        max_index=0
        sum_class1=0
        sum_class2=0
        sum_class3=0
        sum_class4=0
        sum_class5=0
        sum_class6=0
        sum_class7=0
        sum_class8=0
        sum_class9=0
        sum_class10=0
        #check if a string is belong to category one
        for w in l:
            if w in frequences1.keys():
                sum_class1+=frequences1[w]
            else:
                sum_class1+= np.log10( 1/(n_tokens1+n_unique_tokens1))
        predict_list.append(sum_class1+prior_log_list[0])
        #check if a string is belong to category two
        for j in l:
            if j in frequences2.keys():
                sum_class2+=frequences2[j]
            else:
                sum_class2+=np.log10( 1/(n_tokens2+n_unique_tokens2))
        predict_list.append(sum_class2+prior_log_list[1])
        #check if a string is belong to category three
        for x in l:
            if x in frequences3.keys():
                sum_class3+=frequences3[x]
            else:
                sum_class3+=np.log10( 1/(n_tokens3+n_unique_tokens3))
        predict_list.append(sum_class3+prior_log_list[2])
        #check if a string is belong to category four
        for y in l:
            if y in frequences4.keys():
                sum_class4+=frequences4[y]
            else:
                sum_class4+=np.log10( 1/(n_tokens4+n_unique_tokens4))
        predict_list.append(sum_class4+prior_log_list[3])
        #check if a string is belong to category five
        for z in l:
            if z in frequences5.keys():
                sum_class5+=frequences5[z]
            else:
                sum_class5+=np.log10( 1/(n_tokens5+n_unique_tokens5))
        predict_list.append(sum_class5+prior_log_list[4])
        #check if a string is belong to category six
        for r in l:
            if r in frequences6.keys():
                sum_class6+=frequences6[r]
            else:
                sum_class6+=np.log10( 1/(n_tokens6+n_unique_tokens6))
        predict_list.append(sum_class6+prior_log_list[5])
        #check if a string is belong to category seven
        for t in l:
            if t in frequences7.keys():
                sum_class7+=frequences7[t]
            else:
                sum_class7+=np.log10( 1/(n_tokens7+n_unique_tokens7))
        predict_list.append(sum_class7+prior_log_list[6])
        #check if a string is belong to category eight
        for c in l:
            if c in frequences8.keys():
                sum_class8+=frequences8[c]
            else:
                sum_class8+=np.log10( 1/(n_tokens8+n_unique_tokens8))
        predict_list.append(sum_class8+prior_log_list[7])
        #check if a string is belong to category nine
        for n in l:
            if n in frequences9.keys():
                sum_class9+=frequences9[n]
            else:
                sum_class9+=np.log10( 1/(n_tokens9+n_unique_tokens9))
        predict_list.append(sum_class9+prior_log_list[8])
        #check if a string is belong to category ten
        for m in l:
            if m in frequences10.keys():
                sum_class10+=frequences10[m]
            else:
                sum_class10+=np.log10( 1/(n_tokens10+n_unique_tokens10))
        predict_list.append(sum_class10+prior_log_list[9])

        #print("the list is:",predict_list)
        max_index = predict_list.index(max(predict_list))
        list_maxIndex.append(max_index+1)
        #print("This sentence belong to",max_index+1)

    return max_index,list_maxIndex   
```
